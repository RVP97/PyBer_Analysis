```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
plt.xlim(0,5)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, color="green", marker="o", linewidth=2, label='Boston')
ax.legend()
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_ylim(0, 45)
ax.grid()
ax.set_title("PyBer Fare by Month")
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[x * 5 for x in y_axis], c="skyblue", linewidths=2, edgecolors="black", label="Boston")
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare")
ax.set_ylabel("Date")
ax.set_xlim(0,50)
ax.legend()
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.pie(y_axis, labels=x_axis, explode=[.2 if x == max(y_axis) else 0 for x in y_axis ], autopct="%.1f%%", colors=colors)
plt.subplots(figsize=(8, 8))
plt.show()
```
